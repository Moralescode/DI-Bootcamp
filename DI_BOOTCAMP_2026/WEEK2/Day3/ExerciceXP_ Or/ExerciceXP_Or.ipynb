### Explication des différences

La **distribution normale univariée** décrit la probabilité d'une seule variable aléatoire. Visuellement, elle est représentée par une courbe en cloche (comme le montre l'histogramme). Les principaux paramètres sont la moyenne (le centre de la cloche) et l'écart-type (qui détermine la largeur de la cloche).

La **distribution normale multivariée**, quant à elle, généralise la distribution normale à plusieurs variables. Dans notre cas (bivarié), elle décrit la probabilité conjointe de deux variables. Au lieu d'une simple courbe, sa densité de probabilité forme une surface en cloche en 3D. Le nuage de points que nous avons tracé est une projection de cette distribution sur un plan 2D.

**Différences clés :**
- **Dimensionnalité :** Univariée = une variable ; Multivariée = plusieurs variables.
- **Paramètres :** La distribution multivariée nécessite un vecteur de moyennes (une moyenne pour chaque variable) et une matrice de covariance. Cette matrice est cruciale car elle décrit non seulement la variance de chaque variable (sur la diagonale) mais aussi la covariance entre les paires de variables (hors de la diagonale). Une covariance non nulle indique que les variables sont corrélées.
- **Visualisation :** Une distribution univariée est une courbe. Une distribution bivariée (un cas de multivariée) est une surface ou, comme nous l'avons fait, un nuage de points où la densité des points reflète la probabilité.

### Exercice 3 : Tests d'hypothèses statistiques

##### Tâche : Effectuer et interpréter un test ANOVA à l’aide de SciPy pour comparer plus de deux groupes.

##### Groupes à utiliser : Vous pouvez utiliser les données fournies par le code suivant :

```python
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(0)
# Sales data across different regions
region1 = np.random.normal(20000, 3000, 30)  # Region 1 sales
region2 = np.random.normal(22000, 3500, 30)  # Region 2 sales
region3 = np.random.normal(25000, 5000, 30)  # Region 3 sales
sales_data = pd.DataFrame({
    'Region 1': region1,
    'Region 2': region2,
    'Region 3': region3
})
```

```python
from scipy.stats import f_oneway

statistic, pvalue = f_oneway(sales_data['Region 1'], sales_data['Region 2'], sales_data['Region 3'])

print(f'Statistique F : {statistic}')
print(f'P-value : {pvalue}')
```

### Interprétation des résultats

**Statistique F :** C'est le ratio de la variance entre les groupes sur la variance à l'intérieur des groupes. Une valeur F élevée suggère que la variation entre les groupes est plus grande que la variation à l'intérieur des groupes.

**P-value :** C'est la probabilité d'observer une statistique F au moins aussi extrême que celle calculée, en supposant que l'hypothèse nulle (H0) est vraie. Un seuil de signification commun (alpha) est de 0.05.

- Si **p-value < 0.05**, nous rejetons l'hypothèse nulle. Cela signifie qu'il y a une différence statistiquement significative dans les ventes moyennes d'au moins une région par rapport aux autres.
- Si **p-value >= 0.05**, nous ne parvenons pas à rejeter l'hypothèse nulle. Cela signifie qu'il n'y a pas de preuve suffisante pour conclure à une différence significative dans les ventes moyennes entre les régions.

**Pente (slope) :** Pour chaque heure d'étude supplémentaire, le score au test augmente en moyenne de la valeur de la pente. Une pente positive indique une relation positive.
