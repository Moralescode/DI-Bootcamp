## Exercice 2 : Statistiques descriptives
### Tâche : 

Étant donné un ensemble de données d'exemple, calculez la moyenne, la médiane, la variance et l'écart type à l'aide de SciPy.

##### Exemple de jeu de données :

```python
# Exemple de jeu de données fourni pour l'exercice de statistiques descriptives
# Nous utilisons cette liste pour calculer la moyenne, la médiane, la variance et l'écart type.
data = [12, 15, 13, 12, 18, 20, 22, 21]
```

#### Réponse

```python
# Import des bibliothèques nécessaires pour les calculs statistiques
import numpy as np
from scipy import stats

# Jeu de données utilisé pour l'exercice
# Il s'agit d'un petit échantillon d'exemples numériques.
data = [12, 15, 13, 12, 18, 20, 22, 21]

# Calcul des statistiques descriptives de base
mean = np.mean(data)  # moyenne arithmétique
median = np.median(data)  # valeur médiane
variance = np.var(data, ddof=1)  # variance de l'échantillon
std_dev = np.std(data, ddof=1)  # écart type de l'échantillon

# Affichage des résultats
print("Données :", data)
print("Moyenne :", mean)
print("Médiane :", median)
print("Variance (échantillon) :", variance)
print("Écart type (échantillon) :", std_dev)

# Résumé détaillé fourni par SciPy
summary = stats.describe(data)
print("\nRésumé SciPy :", summary)
```

##  Exercice 3 :  Comprendre les distributions
#### Tâche :
 Générer une distribution normale à l'aide de SciPy avec une moyenne de 50 et un écart type de 10. Représenter graphiquement cette distribution.

##### Exemple de code :

```python
# Import des bibliothèques nécessaires pour créer et afficher la distribution normale
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Définition des paramètres de la distribution normale
mean = 50
std_dev = 10

# Génération d'un vecteur de valeurs autour de la moyenne
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)

# Calcul de la densité de probabilité pour chaque valeur x
y = norm.pdf(x, loc=mean, scale=std_dev)

# Tracé de la courbe de densité
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Densité de probabilité')
plt.fill_between(x, y, alpha=0.2)
plt.title('Distribution normale N(50, 10)')
plt.xlabel('Valeur')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)
plt.show()
```

## Exercice 4 :  Application du test t

#### Tâche : Effectuer un test T sur deux ensembles de données générées aléatoirement.

Exemple de code :

```python
# Import des bibliothèques requises pour générer les données et effectuer le test t
import numpy as np
from scipy.stats import ttest_ind

# Génération de deux séries de données aléatoires suivant des lois normales
# data1 est centrée sur 50 et data2 est centrée sur 60.
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Application du test t indépendant pour comparer les deux distributions
stat, p_value = ttest_ind(data1, data2, equal_var=False)

# Affichage des résultats du test
print("T-statistic :", stat)
print("P-value :", p_value)

# Interprétation statistique en fonction du seuil 0,05
if p_value < 0.05:
    print("Le résultat est statistiquement significatif (différence probable entre les deux groupes).")
else:
    print("Le résultat n'est pas statistiquement significatif (pas de preuve solide de différence).")
```

### Exercice 5 : Analyse de régression linéaire
#### Objectif : Appliquer la régression linéaire à un ensemble de données et interpréter les résultats.

#### Tâche : 
Étant donné un ensemble de données de prix de logements ( house_prices) et leurs tailles correspondantes ( house_sizes), utilisez la régression linéaire pour prédire le prix d'une maison en fonction de sa taille.

##### Ensemble de données :
house_sizes: [50, 70, 80, 100, 120] (en mètres carrés)
house_prices: [150 000, 200 000, 210 000, 250 000, 280 000] (en unités monétaires)

#### Questions :
   #### Quelles sont la pente et l'ordonnée à l'origine de la droite de régression ?
   #### Estimez le prix d'une maison de 90 mètres carrés.
   #### Interprétez la signification de la pente dans le contexte des prix de l'immobilier.


```python
# Import de NumPy pour la manipulation de tableaux numériques
import numpy as np

# Données d'entraînement : tailles de maisons et prix correspondants
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Calcul des coefficients de la droite de régression linéaire (pente et intercept)
slope, intercept = np.polyfit(house_sizes, house_prices, 1)

# Prédiction du prix pour une maison de 90 m²
pred_90 = slope * 90 + intercept

# Affichage des résultats
print("Pente (slope) :", slope)
print("Ordonnée à l'origine (intercept) :", intercept)
print("Prix estimé pour une maison de 90 m² :", round(pred_90, 2))

# Explication de la pente dans le contexte des prix de l'immobilier
print("\nInterprétation :")
print("La pente représente l'augmentation moyenne du prix par mètre carré.")
print("Ainsi, chaque m² supplémentaire ajoute environ", round(slope, 2), "unités monétaires au prix.")
```

### Exercice 6 : Comprendre l’ANOVA
#### Objectif : Tester la compréhension de l'ANOVA et de son application.

##### Tâche : Trois engrais différents sont appliqués à trois groupes de plantes distincts afin de tester leur efficacité. La croissance, exprimée en centimètres, est enregistrée.

#### Ensemble de données :



##### fertilizer_1: [5, 6, 7, 6, 5]
##### fertilizer_2: [7, 8, 7, 9, 8]
##### fertilizer_3: [4, 5, 4, 3, 4]


#### Questions :

##### Effectuez un test ANOVA sur les données fournies. Quelles sont les valeurs F et p ?
##### D'après la valeur P, les engrais ont-ils des effets significativement différents sur la croissance des plantes ?
##### Expliquez ce qui se passerait si la valeur P était supérieure à 0,05.

```python
# Import de la fonction f_oneway de SciPy pour effectuer un test ANOVA
from scipy.stats import f_oneway

# Définitions des trois groupes d'engrais et des mesures de croissance associées
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Exécution du test ANOVA à un facteur pour comparer les trois groupes
f_stat, p_val = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# Affichage des statistiques du test
print("F-statistic :", f_stat)
print("P-value :", p_val)

# Décision d'acceptation ou de rejet de l'hypothèse nulle
if p_val < 0.05:
    print("Les engrais ont des effets significativement différents sur la croissance des plantes.")
else:
    print("Les engrais n'ont pas d'effets significativement différents (pas de preuve suffisante).")

# Explication de l'interprétation de la p-value
print("\nInterprétation :")
print("Si la valeur P était supérieure à 0,05, cela signifierait qu'on ne peut pas rejeter l'hypothèse nulle.")
print("Autrement dit, on ne verrait pas de différence statistiquement significative entre les engrais.")
```

### Exercise 7: Distributions de probabilité (facultatif)

Travailler avec une distribution binomiale : calculer les probabilités pour différents nombres de succès.

Exemple : Calculer la probabilité d'obtenir exactement 5 faces en 10 lancers de pièce.

```python
# Import du module binom de SciPy pour travailler avec la distribution binomiale
from scipy.stats import binom

# Définition des paramètres du test : 10 lancers de pièce, probabilité de succès 0,5
n = 10
p = 0.5
k = 5

# Calcul de la probabilité d'obtenir exactement k succès
prob_exact_5 = binom.pmf(k, n, p)
print(f"Probabilité d'obtenir exactement {k} faces en {n} lancers : {prob_exact_5:.4f}")

# Calcul et affichage des probabilités pour tous les nombres de succès possibles
probs = binom.pmf(range(n + 1), n, p)
print("\nProbabilités de 0 à 10 succès :")
for i, prob in enumerate(probs):
    print(f"{i} succès : {prob:.4f}")
```

### Exercise 8: Coefficients de corrélation (facultatif)

Calculez les coefficients de corrélation de Pearson et de Spearman entre deux variables dans un ensemble de données.

Exemple de code :

data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

```python
# Import de pandas pour structurer les données et de SciPy pour calculer les corrélations
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Création du DataFrame d'exemple avec deux variables à comparer
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

# Calcul des coefficients de corrélation
pearson_corr, pearson_p = pearsonr(data['age'], data['income'])
spearman_corr, spearman_p = spearmanr(data['age'], data['income'])

# Affichage des résultats des corrélations
print(f"Corrélation de Pearson : {pearson_corr:.4f}, P-value : {pearson_p:.4f}")
print(f"Corrélation de Spearman : {spearman_corr:.4f}, P-value : {spearman_p:.4f}")

# Interprétation des résultats
print("\nInterprétation :")
print("Les deux coefficients montrent une forte relation positive entre l'âge et le revenu.")
print("Pearson mesure la corrélation linéaire, Spearman mesure l'association monotone.")
```
